```python
invis = '0, 0' # 0 xscale and yscale to make it invisible
```

```python
# IDs
parts = {
    'core': 0,
    'trigger': 1,
}
```

```python
# level's parts code
txt = f'''
core, 400, 300, 0, {invis}, -1, 0, 0066FF
switch1, 0, -80, 0, 4, 4, e0, 1, FFFFFF
'''
```

### fireworks

```python
FIREWORKS_STARTING_ID = 100

xs = np.random.randint(0, 800, 40)
ys = np.random.randint(0, 600, 40)

singu_ids = []
txt += '\n\n\n'
for idx, (x, y) in enumerate(zip(xs, ys)):
    new_id = FIREWORKS_STARTING_ID+idx
    txt += f'singu, {x-400}, {y-300}, 0, {invis}, e0, {new_id}, 000000'
    txt += '\n'
    singu_ids.append(new_id)
    
append(disableGun(singu_ids))
```

### text

```python
# must be higher than all parts in above sections
TEXT_STARTING_ID = 300

# make all circle parts
_ids = []
txt += '\n\n\n'
for idx, (x, y) in enumerate(merry_christmas_points):
    new_id = TEXT_STARTING_ID+idx
    txt += f'c, {x-400:0.2f}, {y-300:0.2f}, 0, {invis}, e0, {new_id}, FFFFFF'
    txt += '\n'
    _ids.append(new_id)

# helper function to make code more succinct
on_trigger = lambda cmd: append(makeTriggerDeath(parts['trigger'], cmd))

delay = 2
init_time_delay = 30
for idx, id in enumerate(_ids):
    t = init_time_delay + (idx+1)*delay
    
    # change size from 0 (this makes it suddenly appear on-screen)
    grow = setSize(id, 20)
    on_trigger(makeTriggerTime(t, grow[0]))
    on_trigger(makeTriggerTime(t, grow[1]))
    
    # change color, for effect, a split second later
    which_color = 'CC0000' if (idx % 2 == 0) else '00CC00'
    change_color = setColor(id, which_color)
    on_trigger(makeTriggerTime(t+5, change_color))

end_time = init_time_delay+(idx+1)*delay
```

### fireworks: turn on

```python
on_trigger(makeTriggerTime(end_time, enableGun(singu_ids)))
```

### end level

```python
on_trigger(makeTriggerTime(end_time+60, destroy(0)))
```
